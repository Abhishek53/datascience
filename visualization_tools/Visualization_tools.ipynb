```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
print(x)
print(y)
```

```python
plt.plot(x,y, 'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('exponential plot')
plt.show()
```

```python
plt.subplot(1,2,1) #plt.subplot(no of rows, no of columns, plot number)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Exponential plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger plot')
axes2.set_title('Smaller plot')
```

```python
#subplots using object oriented programming
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Incresing')
axes[1].plot(y,x)
axes[1].set_title('Decreasing')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#saving the plots as figure
fig.savefig('subplots.png',dpi=400)
```

```python
#plot legend
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label='x squared')
axes.plot(x,x**3, label='x cubed')
axes.legend(loc=0) #0-10 options
```

```python
#cutomizing the options
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', linewidth=5, alpha=0.5, linestyle='steps') #u can also use rgb hex codes

```

```python
#markers
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='red', lw=5,
          marker='o', markersize=15, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='black') #many markers are availabel
```

```python
#setting the scales
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=4, ls='--')
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
